```python
import os
from PIL import Image
```

```python
month = "X"
date = "X"
p_month = ""
p_date = ""

water = 0
ice = 0
snow = 0
total = 0

folder = '/home/pf/pfstud/Roberto_Internship/Groundtruth/2017_18/semantic_segmentation/moritz_cam1_17-18/choppedBG/labels/'
# folder = '/home/pf/pfstud/Roberto_Internship/Groundtruth/2017_18/semantic_segmentation/Cropped_images_Sihl_cam9_17-18/SegmentationClassPNG/'
f=open("../Groundtruth/2017_18/combinedGT_WebcamAndSentinel/StMoritz_webcamSentinelCombined.txt", "a+")

for filename in sorted(os.listdir(folder)):
    if filename.endswith('.png'):
        print(filename)
        month = filename[-14:-12]
        date = filename[-12:-10]
#         print("Month: {}, Date: {}".format(month, date))
        if date != p_date:
            print("Month: {}, Date: {}, Water: {}, Ice: {}, Snow: {}, Total: {}".format(p_month, p_date, water, ice, snow, total))
            
            if total > 0:
                x = ""
                partial = False
                if water / total > 0.9:
                    x = "w"
#                     print("w")
                elif ice / total > 0.9:
#                     print("i")
                    x = "i"
                elif snow / total > 0.9:
                    x = "s"
#                     print("s")
                elif water / total > 0.3:
                    partial = True
                    x = "mw"
#                     print("mw")
                elif ice / total > 0.3:
                    if partial is True:
                        x += "/"
                    partial = True
                    x += "mi"
                elif snow / total > 0.3:
                    if partial is True:
                        x += "/"
                    x += "ms"
                else:
                    x += "***************************************************************************"
                 
                f.write("{}.{} {}\n".format(date, month, x))   
            p_date = date
            p_month = month
            water = 0
            ice = 0
            snow = 0
            total = 0
            
        im = Image.open(folder+filename)
        for pixel in im.getdata():
            if pixel == 127: # if your image is RGB (if RGBA, (0, 0, 0, 255) or so
                water += 1
                total += 1
            elif pixel == 191:
                ice += 1
                total += 1
            elif pixel == 255:
                snow += 1
                total += 1
            
f.close()          
        
```
