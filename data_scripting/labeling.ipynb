```python
import os, numpy, PIL
from PIL import Image

year = "Segl-Maria/2017"
gt_folder = "../Groundtruth/"+year

imlist = []
# Access all PNG files in directory
allfiles=os.listdir(os.getcwd())
for x in sorted(os.walk(gt_folder, topdown=True)):
    folder = gt_folder
    current_folder = x[0]
    for filename in sorted(os.listdir(current_folder))[9:]:
        if filename.endswith('.jpg'):
#             print(current_folder)
#             print(filename)
            imlist.append(current_folder+"/"+filename)
```

```python
imlist[-100:]
```

```python

# Assuming all images are the same size, get dimensions of first image
w,h=Image.open(imlist[0]).size
N=len(imlist)

# Create a numpy array of floats to store the average (assume RGB images)
arr=numpy.zeros((h,w,3),numpy.float)

# Build up average pixel intensities, casting each image as an array of floats
for im in imlist:
    x = Image.open(im)
    print(x.size[0])
    if x.size[0] == w:
        imarr=numpy.array(x,dtype=numpy.float)
        arr=arr+imarr/N

# Round values in array and cast as 8-bit integer
arr=numpy.array(numpy.round(arr),dtype=numpy.uint8)

# Generate, save and preview final image
out=Image.fromarray(arr,mode="RGB")
out.save("Average.png")
out.show()
```
